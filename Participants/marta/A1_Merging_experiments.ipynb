```python
from cPickle import load
```

```python
reso = 1000000
```

```python
hic_data1 = load_hic_data_from_reads ('results/HindIII/03_filtering/valid_reads12_HindIII.tsv',
                                       resolution = reso)
```

```python
hic_data1.bias = load(open('results/HindIII/04_normalizing/biases_1000000_HindIII.pick'))
```

```python
hic_data2 = load_hic_data_from_reads ('results/MboI/03_filtering/valid_reads12_MboI.tsv',
                                       resolution = reso)
```
